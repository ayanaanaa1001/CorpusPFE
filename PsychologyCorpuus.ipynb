```python
import re

def remove_all_years(input_file, output_file):

    with open(input_file, 'r', encoding='utf-8') as f:
        text = f.read()

    # Pattern for Gregorian years (1900-2099)
    gregorian_pattern = r'\b(19|20)\d{2}\b'

    # Pattern for Hijri years (1000-1500 AH)
    hijri_pattern = r'\b(1[0-4]\d{2}|1500)\b'

    year_pattern = re.compile(f'{gregorian_pattern}|{hijri_pattern}')

    cleaned_text = year_pattern.sub('', text)

    with open(output_file, 'w', encoding='utf-8') as f:
        f.write(cleaned_text)

    print(f" All years removed. Saved to: {output_file}")


remove_all_years('/content/cleaned_rData.txt', '/content/no_years_rData.txt')
```

```python
import re

def remove_all_web_references(input_file, output_file):
    with open(input_file, 'r', encoding='utf-8') as file:
        text_content = file.read()


    web_pattern = re.compile(
        r'(?:https?://|www\.)[^\s]+'  # URLs
        r'|[\w\.-]+@[\w\.-]+'         # Emails
        r'|@[^\s]+'                    # Social handles
        r'|#\w+'                       # Hashtags
    )

    cleaned_text = web_pattern.sub('', text_content)

    with open(output_file, 'w', encoding='utf-8') as file:
        file.write(cleaned_text)

    print(f"All web references removed. Saved to: {output_file}")


remove_all_web_references('/content/no_years_rData.txt', '/content/noLinks_cleaned_rData.txt')
```

```python
import re

def remove_non_arabic(input_file, output_file):
    with open(input_file, 'r', encoding='utf-8') as f:
        text = f.read()

    # Remove non-Arabic words
    arabic_only = re.sub(r'[^\u0600-\u06FF\u0750-\u077F\u08A0-\u08FF\uFB50-\uFDFF\uFE70-\uFEFF\s،؛؟.!0-9]', '', text)

    with open(output_file, 'w', encoding='utf-8') as f:
        f.write(arabic_only)

    print(f"Done! Cleaned file saved as {output_file}")

remove_non_arabic('noLinks_cleaned_rData.txt', 'arabic_only_rData.txt')
```

```python
import json
import string


punctuations = set(string.punctuation) | {'،', '؛', '؟', '«', '»', '“', '”', '‘', '’', '…'}

def clean_word(word):
    """Remove all punctuation from a word"""
    if not isinstance(word, str):
        return word
    return ''.join(char for char in word if char not in punctuations)

def process_json_file(input_filename, output_filename, key="word"):

    try:
        with open(input_filename, 'r', encoding='utf-8') as f:
            data = json.load(f)

        # Process the data structure
        if isinstance(data, list):
            cleaned_data = []
            for item in data:
                if isinstance(item, dict) and key in item:
                    cleaned = clean_word(item[key])
                    if cleaned:  # Only keep non-empty results
                        new_item = item.copy()
                        new_item[key] = cleaned
                        cleaned_data.append(new_item)
                else:
                    cleaned_data.append(item)
        elif isinstance(data, dict):
            cleaned_data = {}
            for k, v in data.items():
                if k == key:
                    cleaned = clean_word(v)
                    if cleaned:
                        cleaned_data[k] = cleaned
                else:
                    cleaned_data[k] = v
        else:
            raise ValueError("Unsupported JSON structure")

        # Save cleaned data
        with open(output_filename, 'w', encoding='utf-8') as f:
            json.dump(cleaned_data, f, ensure_ascii=False, indent=2)

        print(f"Successfully processed {input_filename} -> {output_filename}")

    except Exception as e:
        print(f"Error processing {input_filename}: {str(e)}")
        raise

# Process both files
if __name__ == "__main__":
    # Input and output filenames (same directory)
    files_to_process = [
        ("HumanAnnotation.json", "Human_Annotation.json"),
        ("AiAnnotation.json", "Ai_Annotation.json")
    ]

    for input_file, output_file in files_to_process:
        process_json_file(input_file, output_file)
```

```python
import json

# List of words to split
split_words = {
    "بالاضطراب", "باضطراب", "بالاكتئاب", "بالهَوس", "بالتوحد", "بعسر", "بإعاقة",
    "للإعاقة", "للانتحار", "للاكتئاب", "للاضطراب", "بالانتحار", "للمعالجات",
    "بالعلاج", "لبرامج", "لتحسين", "بمفكِّرة", "بالمعالجة", "فمارس", "بمُضادَّات",
    "بإجراء", "بحزن", "بعدم", "بتخريب", "بالملل", "بالقلق", "بمشاعر", "بطاقة",
    "بثقة", "بالبكاء", "لإيذاء", "بانخفاض", "بخوف", "لمحاولة", "لمحاولات", "بالنحيب",
    "لقلة", "بأوهام", "باجترار", "بإعادة", "بتناول", "بالتفاعل", "بنوبة", "بعدم",
    "برغبة", "للتوتر", "بالاعتماد", "كالخوف", "بالتشتت", "بالهلاوس", "بلبسة",
    "كولاء", "بتأخر", "كالكوابيس", "بالذنب", "للانتهاك", "بالتعب", "لتوتر", "بالخوف",
    "بالإحراج", "بنوبات", "بشراهة", "بالانتفاخ", "لنوبات", "بالخدش", "لاضطراب",
    "للقلق", "للأرق", "بالنهم", "بالاضطراب", "للاضطرابات", "بالأرق", "بالاعتلال",
    "بالتلعثم", "التأتأة", "بالحُبسَة", "بحُبسَة", "بجرح", "لإيذاء", "بإيذاء",
    "كالاكتئاب", "بعدم", "بمتلازمة", "بالسير", "بعُسر", "بنوبات",
    "باضطرابات", "كالاضطراب", "كاضطرابات", "كاضطراب", "بالاضطرابات",
    "باضطراب","بالاكتئاب"
}

def process_json_data(input_data):
    new_data = []
    i = 0
    while i < len(input_data):
        entry = input_data[i]
        word = entry['word'].strip()

        if word in split_words and len(word) > 1 and word[0] in ['ب', 'ل', 'ك']:
            # Check if this is part of an existing entity (not at beginning)
            if i > 0 and input_data[i-1]['tag'].startswith('I-'):
                # Get the entity type from previous tag
                entity_type = input_data[i-1]['tag'].split('-')[1]
                prefix_tag = f"I-{entity_type}"
            else:
                prefix_tag = "O"

            # Split the word
            new_data.append({"word": word[0], "tag": prefix_tag})
            new_data.append({"word": word[1:], "tag": entry['tag']})
        else:
            new_data.append(entry)
        i += 1
    return new_data

# File names
input_files = ['Human_Annotation.json', 'Ai_Annotation.json']
output_filenames = ['final_Human_Annotations.json', 'final_Ai_Annotations.json']

# Processing and saving
for in_file, out_file in zip(input_files, output_filenames):
    print(f"\nProcessing {in_file} -> {out_file}")
    with open(in_file, 'r', encoding='utf-8') as f:
        data = json.load(f)
    processed_data = process_json_data(data)
    with open(out_file, 'w', encoding='utf-8') as f:
        json.dump(processed_data, f, ensure_ascii=False, indent=2)
    print(f"Saved processed data to {out_file}")

print("\nAll files processed and saved successfully!")
```

```python
import json

def compare_annotations(human_file, ai_file, output_file):
    # Load JSON files
    with open(human_file, 'r', encoding='utf-8') as f:
        human_data = json.load(f)
    with open(ai_file, 'r', encoding='utf-8') as f:
        ai_data = json.load(f)

    results = {
        'entries': [],
        'counts': {'TP': 0, 'FP': 0, 'TN': 0, 'FN': 0},
        'metrics': {}
    }

    entity_types = {'DISORDER', 'SYMPTOM', 'TREATMENT'}
    bio_prefixes = ('B-', 'I-')

    for human_entry, ai_entry in zip(human_data, ai_data):
        # Extract word and tags (now using 'word' key)
        word = human_entry.get('word', '')  # <-- FIXED: Changed from 'text' to 'word'
        human_tag = human_entry.get('tag', 'O').upper()
        ai_tag = ai_entry.get('tag', 'O').upper()

        # Determine if human tag is an entity
        is_human_entity = (human_tag in entity_types or
                          human_tag.startswith(bio_prefixes))

        # Classify prediction
        if is_human_entity:
            if ai_tag == human_tag:
                result = 'TP'
            elif ai_tag == 'O':
                result = 'FN'
            else:  # AI predicted a different entity type
                result = 'FP'
        else:  # Human tag is 'O'
            if ai_tag == 'O':
                result = 'TN'
            else:  # AI predicted an entity where none exists
                result = 'FP'

        # Update counts
        results['counts'][result] += 1

        # Store detailed comparison
        results['entries'].append({
            'word': word,  # Now correctly populated
            'human_annotation': human_tag,
            'ai_annotation': ai_tag,
            'prediction': result
        })

    # Calculate metrics
    TP, FP, FN = results['counts']['TP'], results['counts']['FP'], results['counts']['FN']
    precision = TP / (TP + FP) if (TP + FP) > 0 else 0
    recall = TP / (TP + FN) if (TP + FN) > 0 else 0
    f1 = 2 * (precision * recall) / (precision + recall) if (precision + recall) > 0 else 0

    results['metrics'] = {
        'precision': round(precision, 4),
        'recall': round(recall, 4),
        'f1_score': round(f1, 4)
    }

    # Save results
    with open(output_file, 'w', encoding='utf-8') as f:
        json.dump(results, f, indent=2, ensure_ascii=False)

    print(f" Results saved to {output_file}")
    print(f"TP: {results['counts']['TP']}, FP: {results['counts']['FP']}, TN: {results['counts']['TN']}, FN: {results['counts']['FN']}")
    print(f"Precision: {results['metrics']['precision']:.4f}, Recall: {results['metrics']['recall']:.4f}, F1: {results['metrics']['f1_score']:.4f}")

# Example usage
compare_annotations(
    human_file='Human_Annotation.json',
    ai_file='Ai_Annotation.json',
    output_file='comparison_results.json'
)
```

```python
import json
import re
from google.colab import files

def remove_tashkeel(text):

    tashkeel = re.compile(r'[\u064b-\u065f\u0670]')
    return tashkeel.sub('', text)

def extract_medical_entities(data, remove_duplicates=True):
    entities = {
        "DISORDER": {"seen": set(), "items": []},
        "SYMPTOM": {"seen": set(), "items": []},
        "TREATMENT": {"seen": set(), "items": []}
    }

    current_entity = []
    current_type = None

    for entry in data:
        tag = entry.get("tag", "O")
        word = remove_tashkeel(entry.get("word", ""))  # Remove tashkeel from each word

        if tag.startswith("B-"):
            if current_entity:
                _save_entity(current_type, current_entity, entities, remove_duplicates)
            current_type = tag.split("-")[1]
            current_entity = [word]

        elif tag.startswith("I-"):
            entity_type = tag.split("-")[1]
            if entity_type == current_type:
                current_entity.append(word)

        else:
            if current_entity:
                _save_entity(current_type, current_entity, entities, remove_duplicates)
                current_entity = []
                current_type = None

    if current_entity:
        _save_entity(current_type, current_entity, entities, remove_duplicates)

    return {
        "DISORDER": entities["DISORDER"]["items"],
        "SYMPTOM": entities["SYMPTOM"]["items"],
        "TREATMENT": entities["TREATMENT"]["items"]
    }

def _save_entity(entity_type, entity_words, entities, remove_duplicates):
    if entity_type not in entities:
        return

    entity_text = " ".join(entity_words)
    storage = entities[entity_type]

    if not remove_duplicates or entity_text not in storage["seen"]:
        storage["items"].append(entity_text)
        storage["seen"].add(entity_text)

# Configuration
REMOVE_DUPLICATES = True

# Directly use the file (assuming it's already in the Colab environment)
with open('final_Human_Annotations.json', 'r', encoding='utf-8') as f:
    data = json.load(f)

results = extract_medical_entities(data, remove_duplicates=REMOVE_DUPLICATES)

# Save and download results
output_files = {
    "DISORDER": "disorders.txt",
    "SYMPTOM": "symptoms.txt",
    "TREATMENT": "treatments.txt"
}

for entity_type, filename in output_files.items():
    with open(filename, 'w', encoding='utf-8') as f:
        for entity in results.get(entity_type, []):
            f.write(entity + '\n')
    files.download(filename)  # Download the file

# Print summary
print("\nProcessing complete!")
print(f"Unique entities extracted (after tashkeel removal):")
print(f"- Disorders: {len(results['DISORDER'])}")
print(f"- Symptoms: {len(results['SYMPTOM'])}")
print(f"- Treatments: {len(results['TREATMENT'])}")
print(f"\nFiles downloaded: {', '.join(output_files.values())}")
print(f"Duplicate removal: {'ENABLED' if REMOVE_DUPLICATES else 'DISABLED'}")
```
